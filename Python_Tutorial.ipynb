## Aufgabe 1

Also eine lineare Funktion muss her. Funktionen in Python definiert man mit dem Schl√ºsselwort **def**.

Auf **def** folgt der Funktionenname, hier nennen wir unsere Funktion **lin_func**. Diese Funktion nimmt sogenannte Argumente entgegen. Wie viele Argumente bzw. Parameter ben√∂tigen wir, um den Funktionswert einer linearen Funktion an einem bestimmten Punkt zu berechnen?

Genau, es sind drei. Wir schreiben $y = m\cdot x +t$. Wenn wir $y(x)$ berechnen wollen, m√ºssen wir also $(x, m, t)$ kennen.

Zum Ausf√ºhren des Codes die Zelle ausw√§hlen und *Shift + Enter* dr√ºcken.

````python
def lin_func(x,m,t):
    y = m*x + t
    return y
````

Jetzt k√∂nnen wir die Funktion aufrufen und erhalten dann ein Ergebnis zur√ºck, weil wir mit **return y** eine R√ºckmeldung in der Funktion anfordern. Die Funktion meldet sich deshalb mit dem $y$ bei uns, welches wir zuvor in festgelegt haben: **y = m x + t** 

````python
lin_func(2,3,1)
````

Wir erhalten das Ergbnis der Rechnung: $3 \cdot 2 + 1$

## Aufgabe 4

Wir definieren eine neue Funktion, diesmal ein Potenzgesetz der Form: $y = a \cdot x^b$

Das l√§uft genau wie in Aufgabe 1.

````python
def power_law(x,a,b):
    
    #in Python schreibt man den Exponenten b als x**b
    y = a* x**b
    return y
````

Jetzt plotten wir das Ganze, ebenfalls analog zur vorherigen Aufgaben. Wir definieren also die Parameter **a** und **b**, sowie eine Liste von x-Werten. Diese Liste **xs** gibt den Wertebereich an, in dem wir plotten.

````python
#define parameters a and b for power law
a = 5
b = -4

#define list of x values
xs = np.linspace(1, 10, 100)

#plot power law
plt.plot(xs, power_law(xs, a, b))

#set title
plt.title("Potenzgesetz")

#show plot
plt.show()
````

## Aufgabe 5

Zeit f√ºr einen doppelt-logarithmischen Graphen, auch hier hilft uns die Bibliothek `matplotlib`:

````python
#plot power law, logarithmically on both scales
plt.loglog(xs, power_law(xs, a, b))

#set title
plt.title("Potenzgesetz doppelt-logarithmisch")

#show plot
plt.show()
````

````python
#magic to make the plot interactive
%matplotlib widget

#get the missing pieces needed for the interactive plot
from matplotlib.widgets import Slider, Button


# The parametrized power law to be plotted
def f(x, a, b):
    return a* x**b

x = np.linspace(1, 10, 1000)

# Define initial parameters
init_a = 5
init_b = -4

# Create the figure and the line that we will manipulate
fig, ax = plt.subplots()
line, = plt.loglog(x, f(x, init_a, init_b), lw=2)
ax.set_xlabel('x in arbitrary units')
ax.set_title(r"Interactive power law: $y = a \cdot x^b$")

# adjust the main plot to make room for the sliders
plt.subplots_adjust(left=0.25, bottom=0.25)

# Make a horizontal slider to control the frequency.
a_amp = plt.axes([0.25, 0.1, 0.65, 0.03])
a_slider = Slider(
    ax=a_amp,
    label='prefactor a',
    valmin=0.1,
    valmax=30,
    valinit=init_a,
)

# Make a vertically oriented slider to control a
b_amp = plt.axes([0.1, 0.25, 0.0225, 0.63])
b_slider = Slider(
    ax=b_amp,
    label="exponent b",
    valmin=-10,
    valmax=10,
    valinit=init_b,
    orientation="vertical"
)


# The function to be called anytime a slider's value changes
def update(val):
    line.set_ydata(f(x, a_slider.val, b_slider.val))
    fig.canvas.draw_idle()


# register the update function with each slider
a_slider.on_changed(update)
b_slider.on_changed(update)

# Create a `matplotlib.widgets.Button` to reset the sliders to initial values.
resetax = plt.axes([0.8, 0.025, 0.1, 0.04])
button = Button(resetax, 'Reset', hovercolor='0.975')


def reset(event):
    a_slider.reset()
    b_slider.reset()
button.on_clicked(reset)


plt.show()
````
