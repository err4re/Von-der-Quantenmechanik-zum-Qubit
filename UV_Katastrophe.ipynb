# UV-Katastrophe
------

Wie enstand die Lichtquantenhypothese? Versucht man mit der klassischen Elektrodynamik die Strahlung eines schwarzen Körpers zu erklären, so erhält man das folgende Ergebnis: 

$ M(\lambda) d\lambda = \frac{2c\pi}{\lambda^4} k_B T d\lambda $

Die spezifische spektrale Ausstrahlungsdichte ist proportional zu $\lambda^{-4}$.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

from scipy.constants import pi, c, k, h
```

```python
#returns classical (spectral) radiance
#i.e. power emitted per unit emitting area, per steradian, per unit wavelength
def rayleigh_jeans(λ,T):
    
    return (2*pi*c/λ**4) * k * T
```

```python
#plot for wavelengths from 0 to 25μm, and T=1000K
λs = np.linspace(1e-6, 25*1e-6, 1000)
T = 1000
```

```python
figure(figsize=(12, 7), dpi=80)
plt.rcParams['xtick.labelsize']=12
plt.rcParams['ytick.labelsize']=12

plt.plot(λs*1e6, rayleigh_jeans(λs, T))
plt.title("Rayleigh-Jeans law", fontsize=18)
plt.ylabel(r"spektrale spezifische Ausstrahlung in $W/(m^2 \mu m)$", fontsize=14)
plt.xlabel("Wellenlänge in μm", fontsize=14)
plt.show()
```

```python
figure(figsize=(12, 7), dpi=80)

plt.loglog(λs, rayleigh_jeans(λs, T))
plt.title("Rayleigh-Jeans law", fontsize=18)
plt.ylabel(r"spektrale spezifische Ausstrahlung in $W/(m^2 \mu m)$", fontsize=14)
plt.xlabel("Wellenlänge in μm", fontsize=14)
plt.show()
```

# Plancksches Strahlungsgesetz
-----

Der schwarze Körper tauscht Energie mit seiner Umfeld nur in gequantelter Form aus. Er kann nur Pakete von Energie abgeben und aufnehmen, diese können nicht beliebig klein sein! (Aber trotzdem sehr klein $\hbar = 6.626 \cdot 10^{-34} \text{Js}$)

Wir erhalten:

$M(\lambda) d\lambda = \frac{2\pi h c^2}{\lambda^5} \frac{1}{\exp\left(\frac{hc}{\lambda k_B T}\right) -1} d\lambda$

```python
λs = np.linspace(1e-6, 25*1e-6, 1000)
```

```python
def plancks_law(λ, T):
    
    return (2*pi*h*c**2)/λ**5 * 1/(np.exp((h*c/(λ*k*T))) -1)
```

```python
figure(figsize=(12, 7), dpi=80)

plt.plot(λs*1e6, plancks_law(λs, T))
plt.title("Plancksches Strahlungsgesetz", fontsize=18)
plt.ylabel(r"spektrale spezifische Ausstrahlung in $W/(m^2 \mu m)$", fontsize=14)
plt.xlabel("Wellenlänge in μm", fontsize=14)
plt.show()
```

```python
λs = np.linspace(8e-6, 25*1e-6, 1000)
```

```python
figure(figsize=(12, 7), dpi=80)

plt.plot(λs*1e6, plancks_law(λs, T))
plt.plot(λs*1e6, rayleigh_jeans(λs, T), color="red")
plt.title("Plancksches Strahlungsgesetz", fontsize=18)
plt.ylabel(r"spektrale spezifische Ausstrahlung in $W/(m^2 \mu m)$", fontsize=14)
plt.xlabel("Wellenlänge in μm", fontsize=14)
plt.show()
```
